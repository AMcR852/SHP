```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,5)

file = 'trans2.TXT'

laser_data = np.loadtxt(file, usecols = 0, skiprows = 2)
num_points = np.size(laser_data)
points = np.arange(1,num_points+1,1)
minval = np.max(laser_data)
calibrated_data = (laser_data - minval)*-1

A = calibrated_data[63000:65000]
B = calibrated_data[92500:94500]
C = np.concatenate([A, B])
num_points2 = np.size(C)
points2 = np.arange(1,num_points2+1,1)

plt.plot(points2[:1000]/10, C[:1000], label = '1st Steady State')
plt.plot(points2[1000:3000]/10, C[1000:3000], label = '2nd Steady State')
plt.plot(points2[3000:4000]/10, C[3000:4000], label = '3rd Steady State')
plt.xlabel('Time (ms)')
plt.ylabel('Displacement from Closed (mm)')
plt.title("Transitions Between Steady States")
plt.legend()
plt.show()
```
