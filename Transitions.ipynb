```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,5)

file = 'trans2.TXT'

laser_data = np.loadtxt(file, usecols = 0, skiprows = 2)
up_data = np.loadtxt(file, usecols = 1, skiprows = 2)
down_data = np.loadtxt(file, usecols = 2, skiprows = 2)
num_points = np.size(laser_data)
points = np.arange(1,num_points+1,1)
minval = np.max(laser_data)

laser_calibrated = (laser_data - minval)*-1
up_calibrated = up_data*2.5
down_calibrated = down_data/0.27

A = laser_calibrated[63000:65000]
B = laser_calibrated[92500:94500]
C = np.concatenate([A, B])
num_points2 = np.size(C)
points2 = np.arange(1,num_points2+1,1)

D = up_calibrated[63000:65000]
E = up_calibrated[92500:94500]
F = np.concatenate([D, E])

G = down_calibrated[63000:65000]
H = down_calibrated[92500:94500]
I = np.concatenate([G, H])

plt.plot(points2[:1000]/10, C[:1000], label = '1st Steady State')
plt.plot(points2[1000:3000]/10, C[1000:3000], label = '2nd Steady State')
plt.plot(points2[3000:4000]/10, C[3000:4000], label = '3rd Steady State')
plt.xlabel('Time (ms)')
plt.ylabel('Displacement from Closed (mm)')
plt.title("Transitions Between Steady States: Laser Displacement")
plt.legend()
plt.show()

plt.plot(points2[:1000]/10, F[:1000], label = '1st Steady State')
plt.plot(points2[1000:3000]/10, F[1000:3000], label = '2nd Steady State')
plt.plot(points2[3000:4000]/10, F[3000:4000], label = '3rd Steady State')
plt.xlabel('Time (ms)')
plt.ylabel('Upstream Pressure (kPa)')
plt.title("Transitions Between Steady States: Upstream Pressure")
plt.legend()
plt.show()

plt.plot(points2[:1000]/10, I[:1000], label = '1st Steady State')
plt.plot(points2[1000:3000]/10, I[1000:3000], label = '2nd Steady State')
plt.plot(points2[3000:4000]/10, I[3000:4000], label = '3rd Steady State')
plt.xlabel('Time (ms)')
plt.ylabel('Upstream Pressure (kPa)')
plt.title("Transitions Between Steady States: Downstream Pressure")
plt.legend()
plt.show()
```
