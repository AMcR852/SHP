```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,5)
name_dict = {
    0: "1st Steady State",
    1: "2nd Steady State",
    2: "3rd Steady State",
    3: "4th (New) Steady State",
    4: " 2"
    5: "3"
}

files = ['state11.TXT', 'state12.TXT', 'state13.TXT', 'state14.TXT', 'state1.TXT','state2.TXT' ]
laser_data_list = []

for i in range(len(files)):
    laser_data = np.loadtxt(files[i], usecols = 0, skiprows = 2)
    laser_data_list.append(laser_data)
    num_points = np.size(laser_data)
    points = np.arange(1,num_points+1,1)
    minval = np.max(laser_data_list[i][:1000])
    calibrated_data = (laser_data_list[i][:1000] - minval)*-1
    plt.plot(points[:1000]/10, calibrated_data)
    plt.xlabel('Time (ms)')
    plt.ylabel('Displacement from Closed (mm)')
    plt.title(name_dict[i])
    plt.show()
```
