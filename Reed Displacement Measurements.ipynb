```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,5)
name_dict = {
    0: "Reed at 0mm",
    1: "Reed at 2mm",
    2: "Reed at 4mm",
    3: "Reed at 6mm",
    4: "Reed at 8mm",
    5: "Reed at 10mm",
    6: "Reed at 12mm"
}

files = ['reednew1.TXT', 'reednew2.TXT', 'reednew3.TXT', 'reednew4.TXT', 'reednew5.TXT','reednew6.TXT','reednew7.TXT']
laser_data_list = []

for i in range(7):
    laser_data = np.loadtxt(files[i], usecols = 0, skiprows = 2)
    laser_data_list.append(laser_data)
    num_points = np.size(laser_data)
    points = np.arange(1,num_points+1,1)
    minval = np.max(laser_data_list[i][:1000])
    calibrated_data = (laser_data_list[i][:1000] - minval)*-1
    plt.plot(points[:1000]/10, calibrated_data)
    plt.xlabel('Time (ms)')
    plt.ylabel('Displacement from Closed (mm)')
    plt.title(name_dict[i])
    plt.show()
```
