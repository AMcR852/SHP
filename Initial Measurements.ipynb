```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,5)
file = 'meas_5.TXT'

laser_data = np.loadtxt(file, usecols = 0, skiprows = 2)
up_data = np.loadtxt(file, usecols = 1, skiprows = 2)
down_data = np.loadtxt(file, usecols = 2, skiprows = 2)
num_points = np.size(laser_data)
points = np.arange(1,num_points+1,1)

# laser: 1V/mm      ----> already in mm, but adjust for where the closed position is and invert
# up:    2.5kPa/V   ----> multiply array by 2.5 for reading in kPa
# down:  0.27mV/Pa  ----> divide array by 0.27 for reading in kPa

laser_calibrated = (laser_data[:1000] - np.max(laser_data[:1000]))*-1
up_calibrated = up_data*2.5
down_calibrated = down_data/0.27

plt.plot(points[:1000],up_calibrated[:1000])
plt.title('Upstream pressure')
plt.xlabel('Scan number')
plt.ylabel('Pressure (kPa)')
plt.show()

plt.plot(points[:1000],down_calibrated[:1000])
plt.title('Downstream pressure')
plt.xlabel('Scan number')
plt.ylabel('Pressure (kPa)')
plt.show()

plt.plot(points[:1000],laser_calibrated[:1000])
plt.title('Reed Displacement')
plt.xlabel('Scan number')
plt.ylabel('Displacement (mm)')
plt.show()

```
